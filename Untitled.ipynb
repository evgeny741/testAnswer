### Объяснение некоторых моментов
#### Во-первых, давайте выберем простое число: p = 17
#### создадим генератор "g", который определяет циклическую группу, и который возвращает уникальные значения для каждого показателя между "1" и "p-1", где "p-1" называется порядком циклической группы. 
#### h - еще один случайно выбранный генератор простого порядка q.

```python
from random import randint
p = 19
q = 7
g = [x for x in range(1, p) if len(set([x**i % p for i in range(1, p)])) == p-1][0]
s = randint(1, q-1)
h = g**s % p
print(h)
```

#### Пегги фиксирует сообщение m, генерирует случайное r и освобождает обязательство c:

```python
m = 45465465
r = randint(1, q-1)
c = ((g**m) * (h**r)) % p
print(c)
```

#### Позже она показывает (m, r) кортеж Виктору, чтобы проверить обязательство.

```python
check = (g**m * h**r) % p
print("Peggy won!") if check == c else "Try again!"
```
